```python
# TASK 1.1: Develop a *Line chart* using the functionality of pandas to show how automobile sales fluctuate from year to year. 

import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv("C:\\Users\\EB521VC\\Downloads\\historical_automobile_sales.csv")

# Line chart for sales fluctuation
data.groupby('Year')['Automobile_Sales'].sum().plot(kind='line')
plt.title('Automobile Sales Fluctuation')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
# Task 1.2 - Plot different lines for categories of vehicle type and analyse the trend to answer the question "Is there a noticeable difference in sales trends between different vehicle types during recession periods?"
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\EB521VC\\Downloads\\historical_automobile_sales.csv")
# Filter the data for recession periods
recession_data = df[df['Recession'] == 1]
# Plot different lines for categories of vehicle type during recession periods
plt.figure(figsize=(12, 6))
for vehicle_type in recession_data['Vehicle_Type'].unique():
    vehicle_data = recession_data[recession_data['Vehicle_Type'] == vehicle_type]
    plt.plot(vehicle_data['Year'], vehicle_data['Automobile_Sales'], label=vehicle_type)
# Add labels and title
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.title('Sales Trends for Different Vehicle Types during Recession Periods')
plt.legend()
# Show the plot
plt.show()
```

```python
# Task 1.3 - Use the functionality of Seaborn Library to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load the historical automobile sales data
df = pd.read_csv("C:\\Users\\EB521VC\\Downloads\\historical_automobile_sales.csv")
# Filter the data for recession and non-recession periods
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]
# Create a Seaborn line plot to compare the sales trend per vehicle type for recession and non-recession periods
plt.figure(figsize=(14, 7))
# Plot for recession period
sns.lineplot(data=recession_data, x='Year', y='Automobile_Sales', hue='Vehicle_Type', marker='o')
# Plot for non-recession period
sns.lineplot(data=non_recession_data, x='Year', y='Automobile_Sales', hue='Vehicle_Type', marker='x', linestyle='--')
# Add labels and title
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.title('Sales Trend per Vehicle Type: Recession vs Non-Recession Periods')
plt.legend(title='Vehicle Type')
# Show the plot
plt.show()
```

```python
# Task 1.4 - Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
import pandas as pd
import matplotlib.pyplot as plt
# Load the historical automobile sales data
df = pd.read_csv("C:\\Users\\EB521VC\\Downloads\\historical_automobile_sales.csv")
# Filter the data for recession and non-recession periods
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]
# Create subplots to compare the variations in GDP during recession and non-recession periods
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 10))
# Plot for recession period
axes[0].plot(recession_data['Year'], recession_data['GDP'], marker='o', linestyle='-', color='b')
axes[0].set_title('GDP Variations during Recession Periods')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('GDP')
# Plot for non-recession period
axes[1].plot(non_recession_data['Year'], non_recession_data['GDP'], marker='x', linestyle='--', color='r')
axes[1].set_title('GDP Variations during Non-Recession Periods')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('GDP')
# Adjust layout
plt.tight_layout()
# Show the plots
plt.show()
```

```python
# Task 1.5 - Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.

import pandas as pd
import matplotlib.pyplot as plt

# Load the historical automobile sales data
df = pd.read_csv("C:\\Users\\EB521VC\\Downloads\\historical_automobile_sales.csv")
# Create a Bubble plot for displaying the impact of seasonality on Automobile Sales
plt.figure(figsize=(14, 7))
plt.scatter(df['Month'], df['Automobile_Sales'], s=df['Seasonality_Weight']*1000, alpha=0.5)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.title('Impact of Seasonality on Automobile Sales')
# Show the plot
plt.show()
```

```python
# Task 1.7 - Create a pie chart to display the portion of advertising expenditure of XYZAutomotives during recession and non-recession periods.

import pandas as pd
import matplotlib.pyplot as plt
# Load the historical automobile sales data
df = pd.read_csv("C:\\Users\\EB521VC\\Downloads\\historical_automobile_sales.csv")
# Calculate the total advertising expenditure during recession and non-recession periods
recession_ad_expenditure = df[df['Recession'] == 1]['Advertising_Expenditure'].sum()
non_recession_ad_expenditure = df[df['Recession'] == 0]['Advertising_Expenditure'].sum()

# Create a pie chart to display the portion of advertising expenditure during recession and non-recession periods
labels = ['Recession', 'Non-Recession']
sizes = [recession_ad_expenditure, non_recession_ad_expenditure]
colors = ['#ff9999','#66b3ff']
explode = (0.1, 0)  # explode the 1st slice (Recession)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Advertising Expenditure of XYZAutomotives during Recession and Non-Recession Periods')
plt.show()
```

```python
# Task 1.8 - Develop a pie chart to display the total Advertisement expenditure for each vehicle type during recession period.

import pandas as pd
import matplotlib.pyplot as plt

# Load the historical automobile sales data
df = pd.read_csv("C:\\Users\\EB521VC\\Downloads\\historical_automobile_sales.csv")
# Filter the data for recession periods
recession_data = df[df['Recession'] == 1]
# Calculate the total advertisement expenditure for each vehicle type during recession period
ad_expenditure_by_vehicle_type = recession_data.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()

# Create a pie chart to display the total advertisement expenditure for each vehicle type during recession period
plt.figure(figsize=(10, 8))
plt.pie(ad_expenditure_by_vehicle_type, labels=ad_expenditure_by_vehicle_type.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Advertisement Expenditure for Each Vehicle Type during Recession Period')
plt.show()
```

```python
# Task 1.9 -  Develop a line plot to analyse the effect of the unemployment rate on vehicle type and sales during the Recession Period. 

import pandas as pd
import matplotlib.pyplot as plt

# Load the historical automobile sales data
df = pd.read_csv("C:\\Users\\EB521VC\\Downloads\\historical_automobile_sales.csv")
# Filter the data for recession periods
recession_data = df[df['Recession'] == 1]
# Create a line plot to analyze the effect of the unemployment rate on vehicle type and sales during the recession period
plt.figure(figsize=(14, 7))
# Plot for each vehicle type
for vehicle_type in recession_data['Vehicle_Type'].unique():
    vehicle_data = recession_data[recession_data['Vehicle_Type'] == vehicle_type]
    plt.plot(vehicle_data['Year'], vehicle_data['unemployment_rate'], label=f'{vehicle_type} - Unemployment Rate')
    plt.plot(vehicle_data['Year'], vehicle_data['Automobile_Sales'], label=f'{vehicle_type} - Sales')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Effect of Unemployment Rate on Vehicle Type and Sales during Recession Period')
plt.legend()
# Show the plot
plt.show()
```

```python
# Task 2.4 - Creating Callbacks; Define the callback function to update the input container based on the selected statistics and the output container.

import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px
# Load the historical automobile sales data
df = pd.read_csv("C:\\Users\\EB521VC\\Downloads\\historical_automobile_sales.csv")
# Initialize the Dash app
app = dash.Dash(__name__)

# Layout of the app
app.layout = html.Div([
    html.H1('Automobile Sales Dashboard'),
    html.Label('Select Vehicle Type:'),
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=[{'label': vt, 'value': vt} for vt in df['Vehicle_Type'].unique()],
        value=df['Vehicle_Type'].unique()[0]
    ),
    html.Label('Select Year:'),
    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': year, 'value': year} for year in df['Year'].unique()],
        value=df['Year'].unique()[0]
    ),
    html.Div(id='output-container', className='output-container')
])

# Callback function to update the input container based on the selected statistics and the output container
@app.callback(
    Output('output-container', 'children'),
    [Input('vehicle-type-dropdown', 'value'),
     Input('year-dropdown', 'value')]
)
def update_output(selected_vehicle_type, selected_year):
    filtered_data = df[(df['Vehicle_Type'] == selected_vehicle_type) & (df['Year'] == selected_year)]
    fig = px.line(filtered_data, x='Month', y='Automobile_Sales', title=f'Sales Trend for {selected_vehicle_type} in {selected_year}')
    return dcc.Graph(figure=fig)

# Run the app
if __name__ == '__main__':
    app.run(debug=True)
```

```python
# Task 2.6 - Create and display graphs for Yearly Report Statistics.

import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv("C:\\Users\\EB521VC\\Downloads\\historical_automobile_sales.csv")

# Group the data by year and sum the automobile sales for each year
yearly_sales = data.groupby('Year')['Automobile_Sales'].sum()

# Create a bar chart for yearly automobile sales
plt.figure(figsize=(10, 5))
yearly_sales.plot(kind='bar', color='blue')
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.title('Yearly Automobile Sales - Bar Chart')
plt.show()

# Create a line graph for yearly automobile sales
plt.figure(figsize=(10, 5))
yearly_sales.plot(kind='line', marker='o', color='green')
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.title('Yearly Automobile Sales - Line Graph')
plt.show()

# Create a pie chart for yearly automobile sales
plt.figure(figsize=(10, 5))
yearly_sales.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Yearly Automobile Sales - Pie Chart')
plt.ylabel('')  # Hide the y-label
plt.show()
```
